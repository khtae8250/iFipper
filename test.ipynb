```python
dele = [7.35, 7.35, 7.09, 6.95, 6.89]
eriksen = [7.45, 7.53, 7.4, 7.08, 6.51]
son = [6.53, 7.14, 6.99, 7.1, 7.23]
kane = [7.53, 7.68, 7.6, 7.38, 7.2]
xs = [2015, 2016, 2017, 2018, 2019]

plt.plot(xs, dele, 'r-', label='Dele Alli')  # x, y,     (r: , -: )
plt.plot(xs, eriksen, 'g--', label='Christian Eriksen')  # (g: , --: )
plt.plot(xs, son, 'b-.', label='Heung-Min Son')  # (b:  -.: 1 )
plt.plot(xs, kane, 'k:', label='Harry Kane')  # (k: , -: )
plt.xlim(2014.5, 2019.5)  # x   (start, end)
plt.ylim(6, 8)  # y   (start, end)
plt.legend(loc="upper right", ncol=1)
plt.xlabel("Year")
plt.ylabel("Rating")
plt.title("DESK's Annual rating")
plt.show()
```
